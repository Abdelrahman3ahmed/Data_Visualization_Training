```python
app.layout=html.Div(children = [
    
    html.Div(children = [
        dcc.Input(id='my_in', value= None , type = 'text'),
        html.Div(id = 'my_div')
        
    ] )
    
    
    
    
    
    
    
    
    
])
```

```python
from dash.dependencies import Input , Output
```

```python
@app.callback(
        Output(component_id= 'my_div', component_property = 'children'),
        Input(component_id='my_in' , component_property = 'value')
)
def update_My_Div(input_val):
    if input_val == None or input_val == '':
        return 'hello'
    else:
        return 'Hi {}'.format(input_val)
```

```python
app.run_server()
```
